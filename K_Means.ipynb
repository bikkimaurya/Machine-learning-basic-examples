```python
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\abhishek maurya\\Desktop\\College.csv")
```

```python
df.info()
df.set_index("Unnamed: 0")
```

```python
from sklearn.cluster import KMeans
new_df=df.drop("Private",axis=1)
#this line is for dropping "Private " column
#here we will not split in train and test becouse it is un superwised learning
```

```python
X=df.iloc[:,2:].values
```

```python
d=KMeans(n_clusters=2)
```

```python
d.fit(X)
```

```python
import matplotlib.pyplot as plt
```

```python
pred=d.predict(X)
```

```python
y=(df[["Private"]].values=="No")+0
```

```python
y
```

```python
import seaborn as sn
from sklearn import metrics
cm=metrics.confusion_matrix(y,pred)
sn.heatmap(cm,annot=True)
print("Accuracy:",metrics.accuracy_score(y,pred))
print("Precision:",metrics.precision_score(y,pred))
print("Recall:",metrics.recall_score(y,pred))
cm
```
