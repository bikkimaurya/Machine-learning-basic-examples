```python
#rename the no-show column to output label
df = df.rename(columns = {"No-show":"Output-Label"})
df.columns
```

```python
#convert yes and no in 1 and 0
df['Output-Label'] = df['Output-Label'].replace({'No':0,'Yes':1})
```

```python
df['City '] = df['City '].replace({'No':0,'Yes':1})
```

```python
df['Gender'] = df['Gender'].replace({'F':0,'M':1})
```

```python
df.head()
```

```python
#Now calculate the prevalence of the positive class
#make function

def cal_prevalence(y_actual):
    p = sum(y_actual)/len(y_actual)
    return p
```

```python
Out = df["Output-Label"].values
print("Prevalence of the positive Class : %.3f"%cal_prevalence(Out))
#20% people didnt show up
```

```python
#find the unique values
for c in list(df.columns):
    n = df[c].unique()
    #if number of unique values is less than 30, print the values,else print the numbers of
    #unique values
    
    if len(n)<30:
        print(c) #print colummn name
        print(n) #print number of  unique values
    else:
        print(c + ":",len(n),"Unique Values")
```

```python
#In our data  we have ?/ so we want to replace ? with null values
import numpy as np
df = df.replace('?',np.nan)
```

```python
#change to the date format
df["ScheduledDay"] = pd.to_datetime(df["ScheduledDay"],format ="%Y-%m-%d %H:%M:%S" )
df["AppointmentDay"] = pd.to_datetime(df["AppointmentDay"],format ="%Y-%m-%d %H:%M:%S" )
df.head()
```

```python
#find the difference between appointment days annd scheduled days and create a new column
df["num_days"] = (df["AppointmentDay"]-df["ScheduledDay"]).dt.days
df.head()
```

```python
df.head()
```

```python
#convert -1 into 0
df.num_days = np.where(df.num_days<0,0,df.num_days)
```

```python
df.head()
```

```python
df["days_of_week"] = df["ScheduledDay"].dt.dayofweek
df["month"] = df["ScheduledDay"].dt.month
df["week"] = df["ScheduledDay"].dt.week
```

```python
df.head()
```

```python
#column names with integer values
col_names = ["SMS_received","Age","num_days","days_of_week","month","week"]
```

```python
#check any value is null
df[col_names].isnull().sum()
```

```python
df[col_names].fillna(df[col_names].mean(),inplace = True)
```

```python
col_cat = ["Gender",
          "PRIORITY_TYPE_CODE",
          'Index of Multiple Deprivation Decile',
          'Income Decile',
          'Employment Decile',
          'Education and Skills Decile',
          'Health and Disability Decile'
          ]
```

```python
df[col_cat].isnull().sum()
```

```python
df["PRIORITY_TYPE_CODE"] = df["PRIORITY_TYPE_CODE"].fillna(df["PRIORITY_TYPE_CODE"].mode()[0])
```

```python
df.head()
```

```python
df["PRIORITY_TYPE_CODE"].mode()
```

```python
for cols_cat in df[col_cat]:
    df[cols_cat].fillna(df[cols_cat].mode()[0], inplace = True)
#[0] for columns
```

```python
#convert type into string
df[col_cat] = df[col_cat].astype('str')
```

```python
df[col_cat].describe()
```

```python
df.keys()
```

```python
df.drop(["City "],axis=1,inplace = True)
```

```python
df.head()
```

```python
pd.get_dummies(df["PRIORITY_TYPE_CODE"])
```

```python
pd.get_dummies(df["Gender"])
```

```python
#to convert all categorical variables into dummy variables
df_cat = pd.get_dummies(df[col_cat])
```

```python
df_cat.keys()
```

```python
df_cat.head()
```

```python
#Join this data with 
new_df = pd.concat([df[col_names], df_cat],axis =1)
new_df.head()
```

```python
cols_all_cat = list(df_cat.columns)
print("Total number of features: ",len(col_names + cols_all_cat))
```

```python
df.keys()
```

```python
y = df["Output-Label"]
x = new_df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3,random_state = 0)
```

```python
x_train.size
```

```python
#apply k-nearest-neighbours
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 100)
knn.fit(x_train,y_train)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state = 0)
lr.fit(x_train,y_train)
```

```python
#apply random forest classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state = 0)
rf.fit(x_train,y_train)
```

```python
#print accuracy
from sklearn.metrics import accuracy_score,precision_score,recall_score,precision_recall_curve
def Print_Report(model_name,y_test,y_pred):
    print("Accuracy of", model_name,accuracy_score(y_test,y_pred))
    print("Recall of ",model_name,recall_score(y_test,y_pred))
    print("Precision of ",model_name,precision_score(y_test,y_pred))
```

```python
#print report for kneighbors classifier
y_pred = knn.predict(x_test)
Print_Report("KNeighborsClassifier", y_test,y_pred)
```

```python
#print report for Logisitic Regression
y_pred = lr.predict(x_test)
Print_Report("LogisticRegression: ", y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state = 0)
dt.fit(x_train,y_train)
```

```python
df_results = pd.DataFrame({
    "Classifier" : ["KNN","LR","DT","RD"],
    "Accuracy": [accuracy_score(y_test, knn.predict(x_test)),
                accuracy_score(y_test, lr.predict(x_test)),
                accuracy_score(y_test, dt.predict(x_test)),
                accuracy_score(y_test, rf.predict(x_test))],
    "Recall":   [recall_score(y_test, knn.predict(x_test)),
                recall_score(y_test, lr.predict(x_test)),
                recall_score(y_test, dt.predict(x_test)),
                recall_score(y_test, rf.predict(x_test))],
    "Precision":[precision_score(y_test, knn.predict(x_test)),
                precision_score(y_test, lr.predict(x_test)),
                precision_score(y_test, dt.predict(x_test)),
                precision_score(y_test, rf.predict(x_test))]
})
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()

fx1 = fig.add_subplot(311)
fx2 = fig.add_subplot(312)
fx3 = fig.add_subplot(313)

fx1.bar(df_results.Classifier,df_results.Accuracy)
fx2.bar(df_results.Classifier,df_results.Recall)
fx3.bar(df_results.Classifier,df_results.Precision)

plt.show()
```
